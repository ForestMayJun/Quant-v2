## 换手率分布均匀度  分钟级数据

```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
tqdm.pandas()
```

```python
data21_23 = pd.read_hdf('/mnt/datadisk2/aglv/aglv/lab_aglv/data21_23.h5')
data21_23
```

```python
def turn_std(vol:pd.DataFrame):
    def _turn_std(vol_daily:pd.DataFrame):
        vol_daily_percent = vol_daily / vol_daily.sum()
        vol_std = vol_daily_percent.std()

        return vol_std
    
    return vol.groupby(level='Date').progress_apply(_turn_std)
```

```python
vol_21_23 = data21_23['LastVolume'].unstack(level='InstrumentID')
vol_21_23
```

```python
eg1 = vol_21_23.xs('20210104', level='Date')
eg1
```

```python
def turn_std_v2(vol:pd.DataFrame, k=0.9):
    def _turn_std(vol_daily:pd.DataFrame):
        vol_daily_percent = vol_daily / vol_daily.sum()
        return vol_daily_percent[vol_daily_percent > vol_daily_percent.quantile(k)].mean()
    
    return vol.groupby(level='Date').progress_apply(_turn_std)
```

```python
turn_fa_v2 = turn_std_v2(vol_21_23)
```

```python
def destop(factor:pd.DataFrame):
    factor.index = [str(i) for i in factor.index]
    res = factor.shift(1) * is_trading.loc[factor.index, factor.columns]
    res.index.names = ['Date']
    return res
```

```python
vol_21_23_sum = vol_21_23.groupby('Date').progress_apply(lambda x: x.sum())
vol_21_23_sum
```

```python
vol_21_23_r5mean = vol_21_23_sum.rolling(5).mean()
vol_21_23_r5mean
```

```python
def turn_std_v3(vol:pd.DataFrame):
    def _turn_std(vol_daily:pd.DataFrame):
        vol_daily_percent = vol_daily / vol_21_23_r5mean
        vol_std = vol_daily_percent.std()

        return vol_std
    
    return vol.groupby(level='Date').progress_apply(_turn_std)
```

```python
turn_fa_v3 = turn_std_v3(vol_21_23)
turn_fa_v3
```
